```python
import pandas as pd
import IPython.display as ipd

max_corrs = pd.read_csv('max_correlations.csv').drop('Unnamed: 0', axis=1) #remove unnamed column which is added as the frame did not have an index when saved
#this shows the most highly correlated factor or predicted risk value for each factor
ipd.display(max_corrs)

factors_data = pd.read_csv('factors_data.csv').set_index('personId')
print(factors_data.sum())
#shows how many positive results for each factor (i.e. how many patients have experienced the factor)

print('The most commonly exerienced factors are:')
print(factors_data.sum().loc[factors_data.sum() >= 20]) #20 chosen as top 5 included values down to 20, but not all, so parameters were changed to include all 20's
#these represent the most commonly experienced factors in the prior year

data = pd.read_csv('data.csv')
data.columns = data.columns.str.title()
data = data.rename(columns={'Personid':'Person ID'})
data = data.drop('Unnamed: 0', axis=1).set_index('Person ID')
ipd.display(data)
print(data['Sex'].value_counts()) #shows how many people there are of each sex (any personId who contained both male and female labels have previously been labelled 'other' as to not skew or misrepresent data)
print(data['Age Group (Years)'].value_counts()) #shows how many people there are within each age group
```
